Need to import these "toolboxes" so we can do certain functions easily.

```
# used to do math/matrix operations in python
import numpy as np

# tools from numpy toolbox used to create matrices/vectors/scalars with 
# different distributions
from numpy.random import randn, randint, rand  

# used to plot images, graphs, histograms, etc.
import matplotlib.pyplot as plt

# used to do a 3-D plot 
from mpl_toolkits.mplot3d import Axes3D
```

Let's make a function for when we plot things. I chose arbitrarily to call the function ***plot***, but I could've chosen any name. Inside the parenthesis, we tell the function what it should expect to receive. In this case, the only argument is a vector/tensor/matrix.

```
def plot(data):
  '''A function to create a histogram of an incoming distribution called data, 
     which should be a vector, matrix, or tensor, etc.'''
  
  # get the shape of the incoming data
  data_shape = data.shape
  
  # if the data is not a vector (i.e. a matrix or tensor), reshape it to a 
  # vector because matplotlib histograms require vectors.
  if len(data_shape) != 1:
    data = data.flatten()
    
  fig = plt.figure() # initialize a figure
  a1 = fig.add_subplot(111)
  a1.set_ylabel('Frequency')  # provide a label for the y axis in subplot a1
  a1.set_xlabel('Value') # provide a label for the x axis in subplot a1
  a1.hist(data, bins=100)  # plot the histogram and bin the values into 100 different bins
  plt.show()  # show the plot
  
  return  # return back to the main program
```

One of the most important things you'll need to know is how to initialize a tensor/matrix/vector. Below, initialize two matrices, one called ***data*** with shape 1000 x 3 with random values from a normal (Gaussian) distribution, and one that equals ***data*** + 0.5 called ***data2***. We will pretend this is a dataset of 2000 samples and 3 features per sample where data represents one class and data2 represents a different one we want to distinguish between.

```
data = rand(1000, 3)
data2 = data + .5
```

Now, we can use matplotlib to plot these points since there are only 3 features per sample. Any more than three and you cannot really visualize it easily. First, we need to pull out the x-, y-, and z-coordinates from the data (which are the first, second and third columns).

```
x, y, z = data[:, 0], data[:, 1], data[:, 2]
x2, y2, z2 = data2[:, 0], data2[:, 1], data2[:, 2]

fig = plt.figure()
a1 = fig.add_subplot(111, projection='3d')
a1.set_xlabel('x-coordinate')
a1.set_ylabel('y-coordinate')
a1.set_zlabel('z-coordinate')
a1.scatter(x, y, z, c='r')
a1.scatter(x2, y2, z2, c='b')
plt.tight_layout()
plt.show()
```

Create two matrices with ***randn***... matrix ***A*** with shape 3 x 4 and matrix ***B*** with shape 5 x 4. 

```
A = randn(3, 4)
B = randn(5, 4)
```

Transpose matrix ***B***.

```
B = B.T   # can also do B = B.transpose()
```

Print both matrices and their shapes below

```
print(A.shape)
print(B.shape)
print(A)
print(B)
```

Perform a matrix-multiplication between matrix ***A*** and ***B*** and call the product ***C***. Then, print the shape of ***C***.

```
C = np.matmul(A, B)
print(C.shape)
```

Square each number in C.

```
C = C ** 2  # ** used to indicate power
```

Take the first row of ***C*** and call that c_row0. Then print the shape of it and print the actual vector.

```
c_row0 = C[0, :]   # take the 0th row of C
print(c_row0.shape)  # output the shape of this row
print(c_row0)  # output the actual vector
```

Take column 3 (4th column) of ***C*** and call that ***c_col3***. Then print the shape

```
c_col3 = C[:, 3]
print(c_col3.shape)
```

Create a vector of ones with length 5 and call it ***ones_vec*** and print it out. Then, perform an element-wise addition with ***c_row0*** and call the result ***result1***.

```
ones_vec = np.ones([5, ])  # create the vector
print(ones_vec)  # print the vector 

result1 = ones_vec + c_row0
print(result1)
```

Now, perform an element-wise multiplication between ***result1*** and ***c_row0*** and call the result ***result2***.

```
result2 = result1 * c_row0
```

Perform a dot-product between ***result1*** and ***result2*** and print the result. 

```
dot_12 = np.dot(result1, result2)
print(dot_12)
```
