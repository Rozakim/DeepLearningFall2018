So the image appears as a 3-dimensional array with 264 rows, 191 columns, and 3 channels (which indicate it is a color image). Let's plot a histogram of the values, but remember to flatten it first. 

```
plt.hist(turing.flatten(), bins=100)
plt.show()
```

So we can see that the values go from 0 to about 250. Let's find out exactly what the maximum and minimum values are, as well as the mean. 

```
turing_min = np.amin(turing)
turing_max = np.amax(turing)
turing_mean = np.mean(turing)
print(turing_min)
print(turing_max)
print(turing_mean)
```

Often, we want the mean of the image to be zero (or close to it). To do this, we subtract the mean so the data is centered on zero (this works a little differently when we have more than one data example/image, as we will see). 

```
turing_norm = turing - np.mean(turing)
plt.hist(turing_norm.flatten(), bins=100)
plt.show()
```

As you probably know, the values in many images usually ranges from 0 to 255, but sometimes we want them to be within a smaller range, like 0 to 1. One way to transform the data from 0 to 255 (or some other maximum value) to 0 to 1 is to divide by the standard deviation.

```
turing_scaled = turing_norm / np.std(turing_norm)
print(np.amin(turing_scaled))
print(np.amax(turing_scaled))

plt.hist(turing_scaled.flatten(), bins=100)
plt.show()
```

Let's see what the image looks like now that we normalized it. However, matplotlib wants the image values to be between 0 - 1 or 0 - 255 with no negative values, so we can import *bytescale* from matplotlib to transform the data to this range. Of course, we could multiply the standard deviation from before and add the mean from before, but that will reverse the changes we have made. 

```
from scipy.misc import bytescale
plt.imshow(bytescale(turing_scaled, 0, 1))
plt.grid(False)
plt.show()
```

We can also make the image grayscale. There are many different ways to do this, but the most simple is probably to take the mean of all three color channels. Notice how the shape changes when we do this. We will use the original turing image we initially read in to do this.

Finally, it is also sometimes important to know how to take patches from an image or data example. This is mostly used in unsupervised learning methods, but can also be applicable to supervised. We will demonstrate one way to take patches from an image below by using for loops. First the method with the for loops. We will take 15 x 15 patches out of the image. We will store all of these patches in a list called patches, which we initialize as empty before the loops. Then we loop through both the rows and columns of the image and take the row plus 15 pixels, the column plus 15 pixels, and all of the color channels, and add this to the bottom of patches. 

```
rows = turing_gray.shape[0]
cols = turing_gray.shape[1]
patch_sz = 15
patches = []

for m in range(rows-patch_sz):
  for n in range(cols-patch_sz):
    patch = turing[m:m+patch_sz, n:n+patch_sz, :]
    patches.append(patch)
    
```

```
patches = np.asarray(patches)
print(patches.shape)
```

Let's look at 2500 random patches We can arrange them all in a grid, which will have the shape of 50 patches x 50 patches.

```
grid_shape = int(np.sqrt(2500))
print(grid_shape)
patch_grid = np.zeros([grid_shape*patch_sz, grid_shape*patch_sz, 3])

for i in range(grid_shape):
  for j in range(grid_shape):
    patch = patches[np.random.randint(0, patches.shape[0], 1)[0], ...]
    patch_grid[i*patch_sz:i*patch_sz+patch_sz, j*patch_sz:j*patch_sz+patch_sz, :] = patch
```

```
fig = plt.figure(figsize=(20, 20))
a1 = fig.add_subplot(111)
a1.imshow(patch_grid)
a1.grid(False)
plt.show()
```
